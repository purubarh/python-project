```python
barplot_mother_tongue = (alt.Chart(first_top_ten).mark_bar()
                         .encode(x = alt.X("language").title("Language"),
                                 y = alt.Y("mother_tongue").title("Mother Tongue (Number of Canadina Residents)"))
                         
                         )

barplot_mother_tongue
```

```python
# Rotate the Bar chart and consider Mother Tongue as X axis and Language as Y axis
```

```python
barplot_mother_tongue = (alt.Chart(first_top_ten).mark_bar()
                         .encode(x = alt.X("mother_tongue").title("Mother Tongue (Number of Canadina Residents)"),
                                 y = alt.Y("language").title("Language"))
                         
                         )

barplot_mother_tongue
```

```python
# We can reorder the bars using the sort method, 
# which orders a variable (here language) based on the values 
# of the variable(mother_tongue) on the x-axis.
```

```python
ordered_barplot_mother_tongue = alt.Chart(first_top_ten).mark_bar().encode(
    x=alt.X("mother_tongue").title("Mother Tongue (Number of Canadian Residents)"),
    y=alt.Y("language").sort("x").title("Language")
)

ordered_barplot_mother_tongue
```

```python
# Final Code for 
# Which ten Aboriginal languages were most often reported in 2016 as mother tongues 
# in Canada, and how many people speak each of them?

# load the data set can_lang.csv
can_lang = pd.read_csv("data/can_lang.csv")

# Obtain the top 10 most common Aboriginal Languages

top_ten = (
    can_lang.loc[can_lang["category"] == "Aboriginal languages", ["language", "mother_tongue"]]
    .sort_values(by = "mother_tongue", ascending = False)
    .head(10)

)

# create the visualization

top_ten_plot = alt.Chart(top_ten).mark_bar().encode(
    x = alt.X("mother_tongue").title("Mother Tongue (Number of Canadian Residents)"),
    y =  alt.Y("language").sort("x").title("Language")
)

top_ten_plot
```
